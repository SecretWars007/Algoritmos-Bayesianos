# Historia


Las Redes Bayesianas fueron desarrolladas
formalmente en los años 1980.


Uno de sus principales investigadores fue:


Judea Pearl


quien introdujo métodos de razonamiento
probabilístico mediante grafos causales.


Una Bayesian Belief Network combina:


- Teoría de probabilidades.
- Grafos dirigidos.
- Inferencia estadística.


Aplicaciones:


✓ Medicina

✓ Diagnóstico inteligente

✓ Sistemas expertos

✓ Predicción de riesgos

✓ Inteligencia artificial

```python
# ==========================================================
# Librerías
# ==========================================================


import pandas as pd

import numpy as np


import matplotlib.pyplot as plt


from pgmpy.models import (
    DiscreteBayesianNetwork
)


from pgmpy.estimators import (
    MaximumLikelihoodEstimator
)


from pgmpy.inference import (
    VariableElimination
)



print(
    "Librerías cargadas correctamente"
)
```

```python
# ==========================================================
# Crear modelo Bayesian Belief Network
# ==========================================================


model = DiscreteBayesianNetwork(

    [

        ("Fiebre", "Gripe"),

        ("Tos", "Gripe"),

        ("Dolor", "Gripe")

    ]

)



print("="*60)

print("Red Bayesiana creada correctamente")

print("="*60)



print(
    model.edges()
)
```

```python
# ==========================================================
# Visualización Red Bayesiana DAG
# ==========================================================


import networkx as nx


plt.figure(
    figsize=(8,5)
)


G = nx.DiGraph()


G.add_edges_from(
    model.edges()
)


pos = {

    "Fiebre":(-1,1),

    "Tos":(0,1),

    "Dolor":(1,1),

    "Gripe":(0,0)

}



nx.draw(

    G,

    pos,

    with_labels=True,

    node_size=3000,

    font_size=12

)



plt.title(
    "Bayesian Belief Network - DAG"
)


plt.show()
```

```python
# ==========================================================
# Aprendizaje de parámetros Bayesian Network
# Compatible con pgmpy 1.x
# ==========================================================


model.fit(
    data
)


print(
    "Parámetros aprendidos correctamente"
)
```

```python
# ==========================================================
# Mostrar Tablas de Probabilidad Condicional
# ==========================================================


for cpd in model.get_cpds():

    print("="*70)

    print(cpd)
```

```python
# ==========================================================
# Inferencia Bayesiana
# ==========================================================


from pgmpy.inference import VariableElimination



inference = VariableElimination(
    model
)



resultado = inference.query(

    variables=["Gripe"],


    evidence={

        "Fiebre":1,

        "Tos":1,

        "Dolor":1

    }

)



print(resultado)
```

```python
# ==========================================================
# Función predictora BBN
# ==========================================================


def predecir_gripe(
    fiebre,
    tos,
    dolor
):


    resultado = inference.query(

        variables=["Gripe"],


        evidence={

            "Fiebre":fiebre,

            "Tos":tos,

            "Dolor":dolor

        }

    )


    return resultado



# Nuevo paciente

nuevo_paciente = predecir_gripe(

    fiebre=1,

    tos=1,

    dolor=1

)



print(
    nuevo_paciente
)
```

```python
# ==========================================================
# Probabilidad inicial de Gripe
# ==========================================================


prob_gripe = (

    data["Gripe"].sum()

    /

    len(data)

)



print(
    "P(Gripe)=",
    prob_gripe
)
```

# Comparación de Modelos Bayesianos


|Característica|Naive Bayes|Bayesian Belief Network|
|-|-|-|
|Modelo|Probabilístico|Probabilístico|
|Dependencia variables|No|Sí|
|Estructura|Producto probabilístico|Grafo DAG|
|Causalidad|No representa|Puede representar|
|Interpretabilidad|Media|Alta|
|Complejidad|Baja|Mayor|


## Naive Bayes


\[
P(C|X)
=
P(C)
\prod_i P(X_i|C)
\]


Asume independencia.


---

## Bayesian Belief Network


\[
P(X_1,...,X_n)
=
\prod_i P(X_i|Parents(X_i))
\]


Permite relaciones entre variables.
# Conclusiones


La Bayesian Belief Network es un modelo
probabilístico basado en grafos.


Características principales:


✓ Representa relaciones entre variables.

✓ Usa grafos dirigidos acíclicos (DAG).

✓ Aprende probabilidades condicionales.

✓ Permite inferencia sobre nuevos datos.

✓ Integra conocimiento experto.


Aplicaciones:


- Diagnóstico médico.

- Sistemas expertos.

- Inteligencia artificial.

- Análisis financiero.

- Sistemas de recomendación.


Limitaciones:


- Construcción del grafo puede ser compleja.

- Requiere conocimiento del dominio.

- Puede necesitar muchos datos.
